# 05 — Qiskit circuits and entanglement entropy

**Manuscript:** Part IV postulates, Part I decoherence/observation.

Build a small circuit (Bell pairs or entangling gates), get statevector, partial trace over subsystems, compute $S(A)$.

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, DensityMatrix, partial_trace, entropy

# 3-qubit circuit: create entanglement
qc = QuantumCircuit(3)
qc.h(0)
qc.cx(0, 1)
qc.cx(1, 2)
sv = Statevector.from_instruction(qc)
print('Statevector (amplitudes):')
print(sv)
rho = DensityMatrix(sv)
```

```python
# Partial trace: subsystem A = qubit 0
rho_A = partial_trace(rho, [1, 2])  # trace out qubits 1, 2
S_A = entropy(rho_A)
print(f'Subsystem A (qubit 0): S(A) = {S_A:.4f} (base 2)')

# Subsystem B = qubits 0,1
rho_B = partial_trace(rho, [2])
S_B = entropy(rho_B)
print(f'Subsystem B (qubits 0,1): S(B) = {S_B:.4f} (base 2)')
```
